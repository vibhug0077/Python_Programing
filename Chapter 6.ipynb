## 1\. Overview of NumPy

**NumPy** stands for **Numerical Python**. It is a powerful Python library used for:

- numerical computations
    
- scientific computing
    
- working with arrays and matrices
    
- performing mathematical and statistical operations efficiently
    

NumPy is faster than normal Python lists because it is implemented in optimized C code and stores data in contiguous memory locations.

### Why NumPy is important

- supports large multi-dimensional arrays
    
- performs fast calculations
    
- provides many built-in mathematical functions
    
- used in data science, machine learning, AI, scientific programming, and analytics
    

### Features of NumPy

- efficient array object called `ndarray`
    
- less memory usage
    
- faster execution than Python lists
    
- supports vectorized operations
    
- supports broadcasting
    
- has mathematical, statistical, and linear algebra functions
    

### Importing NumPy

```python
import numpy as np
```

# 2\. NumPy ndarray

The main object in NumPy is the **ndarray**, which means **N-dimensional array**.

It is a collection of elements of the same data type arranged in one or more dimensions.

### Examples

- 1D array → like a list
    
- 2D array → like a table or matrix
    
- 3D array → collection of matrices
    

### Example

```python
import numpy as np
a = np.array([1, 2, 3, 4])
print(a)
```

### 2D example

```python
b = np.array([[1, 2], [3, 4]])
print(b)
```

### Advantages of ndarray

- homogeneous data
    
- fast computation
    
- supports multi-dimensional data
    
- easy mathematical operations

# 5\. List vs Array

Python list and NumPy array may look similar, but they are different.

## Python List

- can store different datatypes
    
- slower for numerical operations
    
- consumes more memory
    
- does not support vectorized operations directly
    

## NumPy Array

- stores same datatype elements
    
- faster
    
- memory efficient
    
- supports vectorized mathematical operations
    

## Example

### List


```python
l = [1, 2, 3]
print(l * 2)
```

### Array

```python
a = np.array([1, 2, 3])
print(a * 2)
```

## Main differences

| Basis | Python List | NumPy Array |
| --- | --- | --- |
| Datatype | can be mixed | mostly same datatype |
| Speed | slower | faster |
| Memory | more | less |
| Operations | limited | efficient numerical operations |
| Use | general-purpose | numerical computing |

# 8\. NumPy Broadcasting

**Broadcasting** means NumPy can perform arithmetic operations on arrays of different shapes under certain rules.

It automatically expands smaller arrays to match larger arrays without actually copying data.

## Example

```python

a = np.array([1, 2, 3])
b = 2
print(a + b)
```

## Another example

```python
a = np.array([[1], [2], [3]])
b = np.array([10, 20, 30])
print(a + b)
```

## Broadcasting rules

1.  dimensions must be equal, or
    
2.  one of them must be 1
    

If not, broadcasting gives an error.

### Importance

- avoids loops
    
- faster operations
    
- concise code

Broadcasting is the process by which NumPy performs arithmetic operations on arrays of different shapes by automatically expanding smaller arrays.
