

## 1) Single inheritance: call parent `__init__`

```python
class Pet:
    def __init__(self, name):
        self.name = name
        self.age = 10

class Dog(Pet):
    def __init__(self, name, breed):
        super().__init__(name)   # parent init
        self.breed = breed
```

```python
d = Dog("Bruno", "Labrador")
print(d.name, d.breed)
```

```python
print(d.__dict__)
```

## 2) Override a method but keep parent behavior

```python
class Logger:
    def log(self, msg):
        return f"[LOG] {msg}"

class TimeLogger(Logger):
    def log(self, msg):
        base = super().log(msg)  # call parent method
        return base + " (timestamp added)"
```

```python
t = TimeLogger()
print(t.log("hello"))
```
