# Q2 - Hollow Square of side 'N
Problem Description: You are given an integer n. Your task is to return a hollow square pattern of size n x n made up of the character '*', represented as a list of strings. The hollow square has '*' on the border, and spaces ' ' in the middle (except for side lengths of 1 and 2).

**Input Parameters:**

n (int): The size of the square (number of rows and columns).

**Output**:

A list of strings where each string is a row of n characters, representing a hollow square.

Example:
````
Input: 3
Output: ['***', '* *', '***']
 
Input: 5
Output: ['*****', '*   *', '*   *', '*   *', '*****']
````

``````python
def hollowSquare(n):
    pattern = []
    
    for i in range(n):
        if i == 0 or i == n - 1:
            pattern.append('*' * n)
        else:
            pattern.append('*' + ' ' * (n - 2) + '*')
    
    return pattern
            
            
``````

# Q5 - Pyramid Pattern
**Problem Description:**

You are given an integer n. Your task is to return a pyramid pattern of '*' where each side has n rows, represented as a list of strings. The pyramid is centered, with 1 star in the first row, 3 stars in the second row, and so on, increasing by 2 stars per row until the base row has 2n - 1 stars.

**Input:**

A single integer n, where 1 <= n <= 100.

**Output:**

A list of strings where each string contains stars ('*') centered, forming a pyramid shape. Each row has an increasing number of stars, with appropriate spaces for centering.

Example:
````
Input: 3
Output: ['  *  ', ' *** ', '*****']
 
Input: 5
Output: ['    *    ', '   ***   ', '  *****  ', ' ******* ', '*********']
````

``````python
def pyramid_patter(n):
    pattern = []
    for i in range(n):
        spaces = " " * (n -i -1)
        stars = "*" * (2*i +1)
        row = spaces+stars+spaces
        pattern.append(row)
    return(pattern)
``````

# Q6 - Inverted Pyramid

**Problem Description**

You are given an integer n. Your task is to return an inverted pyramid pattern of '*', where each side has n rows, represented as a list of strings. The first row has 2n - 1 stars, the second row has 2n - 3 stars, and so on, until the last row has 1 star, with each row centered using spaces.

**Input Parameters:**

n (int): The number of rows in the inverted pyramid.

Output:

A list of strings where each string represents a row of the inverted pyramid.

Example:
````
Input: 3
Output: ['*****', ' *** ', '  *  ']
 
Input: 5
Output: ['*********', ' ******* ', '  *****  ', '   ***   ', '    *    ']
````

``````python
def invertedPyramid(n):
    pattern = []
    
    for i in range(n):
        spaces = ' ' * i
        stars = '*' * (2 * (n - i) - 1)
        row = spaces + stars + spaces
        pattern.append(row)
    
    return pattern

``````

# Q12 - Remove Duplicate in a List
**Remove Duplicates from a List**

You are given a list of integers. Write a Python program that removes any duplicate elements from the list and returns a new list with only unique elements. The order of elements in the list should be maintained.

**Parameters:
**
lst (List of integers): The list of integers from which duplicates should be removed.

**Returns:**

A list of integers where all duplicates have been removed, preserving the original order.

Example:
````
Input: lst = [1, 2, 2, 3, 4, 4, 5]
Output: [1, 2, 3, 4, 5]

Input: lst = [4, 5, 5, 4, 6, 7]
Output: [4, 5, 6, 7]
`````

``````python
def remove_duplicates(l):
    result_list =[]
    for each in l:
        found = False
        for x in result_list:
            if x == each:
                found = True
                break
        if not found:
            result_list.append(each)
    return result_list
``````

An integer representing the maximum difference between two consecutive elements.

**Example:**
```
Input: lst = [1, 7, 3, 10, 5]

Input: lst = [10, 11, 15, 3]
Output: 12

The maximum difference is between 15 and 3 (i.e., |15 - 3| = 12).
````

``````python
def find_largest_difference(lst):
    larg = 0
    for each in range(len(lst)-1):
        
        print("Value of larg :", larg)
        if larg < abs(lst[each+1] - lst[each]):
            larg = abs(lst[each+1] - lst[each])
        print("Value of larg after update: ", larg)
    return larg
``````
