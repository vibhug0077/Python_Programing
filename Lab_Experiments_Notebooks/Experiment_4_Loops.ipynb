## Question 1

Find factorial of a given number using a loop.

```python
num = 5
factorial = 1

for i in range(1, num + 1):
    factorial *= i

print(f"Factorial of {num} is {factorial}")
```

## Question 2

Check whether the given number is an Armstrong number.

```python
num = 153
num_str = str(num)
power = len(num_str)
sum_of_powers = 0

for digit in num_str:
    sum_of_powers += int(digit) ** power

if sum_of_powers == num:
    print(num, "is an Armstrong number")
else:
    print(num, "is not an Armstrong number")
```

## Question 3

Print Fibonacci series up to a given number of terms.

```python
terms = 10
a, b = 0, 1

print("Fibonacci series:")
for _ in range(terms):
    print(a, end=" ")
    a, b = b, a + b
print()
```

## Question 4

Check whether a given number is prime or not.

```python
num = 29

if num < 2:
    print(num, "is not prime")
else:
    is_prime = True
    for i in range(2, int(num ** 0.5) + 1):
        if num % i == 0:
            is_prime = False
            break
    print(num, "is prime" if is_prime else "is not prime")
```

## Question 5

Check whether a given number is palindrome or not.

```python
num = 12321

if str(num) == str(num)[::-1]:
    print(num, "is a palindrome")
else:
    print(num, "is not a palindrome")
```

## Question 6

Print sum of digits of a number.

```python
num = 9876
sum_digits = 0

for digit in str(num):
    sum_digits += int(digit)

print("Sum of digits:", sum_digits)
```

## Question 7

Count and print all numbers divisible by 5 or 7 between 1 and 100.

```python
numbers = [n for n in range(1, 101) if n % 5 == 0 or n % 7 == 0]
print("Numbers:", numbers)
print("Count:", len(numbers))
```

Convert all lowercase letters to uppercase in a string.

## Question 9

Print all prime numbers between 1 and 100.

```python
primes = []

for num in range(2, 101):
    is_prime = True
    for i in range(2, int(num ** 0.5) + 1):
        if num % i == 0:
            is_prime = False
            break
    if is_prime:
        primes.append(num)

print("Prime numbers from 1 to 100:")
print(primes)
```
