## Question 1

Count and display the number of capital letters in a given string.

```python
text = "PyThOn LAB 2026"
count_caps = sum(1 for ch in text if ch.isupper())
print("Capital letters:", count_caps)
```

## Question 2

Count total number of vowels in a given string.

```python
text = "Welcome to Python Programming"
vowels = "aeiouAEIOU"
count_vowels = sum(1 for ch in text if ch in vowels)
print("Total vowels:", count_vowels)
```

## Question 3

Input a sentence and print words in separate lines.

```python
sentence = "Python is easy to learn"

for word in sentence.split():
    print(word)
```

## Question 4

Enter a string and substring, then print how many times substring occurs from left to right.

```python
main_string = "ABCDCDC"
sub_string = "CDC"

count = 0
start = 0
while True:
    idx = main_string.find(sub_string, start)
    if idx == -1:
        break
    count += 1
    start = idx + 1

print("Occurrences:", count)
```

## Question 5

Count occurrences of each alphabet (case-insensitive) in a mixed-case string.

```python
text = "ABaBCbGc"
frequency = {}

for ch in text:
    if ch.isalpha():
        key = ch.upper()
        frequency[key] = frequency.get(key, 0) + 1

for key in sorted(frequency):
    print(f"{frequency[key]}{key}")
```

## Question 6

Count number of unique words in a sentence using sets.

```python
sentence = "python is simple and python is powerful"
words = sentence.lower().split()
unique_words = set(words)

print("Unique words:", unique_words)
print("Count of unique words:", len(unique_words))
```

## Question 7

Create two sets of fruits and find:
- Fruits in both sets
- Fruits only in `s1`
- Count of all fruits from both sets

```python
s1 = {"apple", "banana", "mango", "grapes"}
s2 = {"banana", "orange", "kiwi", "mango"}

print("Fruits in both sets:", s1 & s2)
print("Fruits only in s1:", s1 - s2)
print("Count of all unique fruits:", len(s1 | s2))
```

## Question 8

Take two sets and apply various set operations:
- `S1 = {Red, yellow, orange, blue}`
- `S2 = {violet, blue, purple}`

```python
S1 = {"Red", "yellow", "orange", "blue"}
S2 = {"violet", "blue", "purple"}

print("Union:", S1 | S2)
print("Intersection:", S1 & S2)
print("Difference (S1 - S2):", S1 - S2)
print("Difference (S2 - S1):", S2 - S1)
print("Symmetric Difference:", S1 ^ S2)
```
